```python
from __future__ import division, print_function, unicode_literals
import pickle
from embed_tsne_plot import embedTsnePlt
import numpy as np
```

```python
if __name__ == '__main__':
                
    # read python dict back from the file
    #  pkl filedictionary
    reverse_dictionary = {}
    pkl_file = open('reverse_dict.pkl', 'rb')
    reverse_dictionary = pickle.load(pkl_file)
    pkl_file.close()     
        
    glove_embed_load = np.load('glove_embeddings.npy')
    
    em_ts_plt0 = embedTsnePlt(distance_threshold = 0.25,
                 sample_threshold = 10)
    twodembed_t0, sel_ids_t0 = em_ts_plt0.tsne_prune(glove_embed_load,
                                             num_points = 1000)
    words_t0 = [reverse_dictionary[i] for i in sel_ids_t0]    
    em_ts_plt0.tsne_plot(twodembed_t0, words_t0)
    
    em_ts_plt1 = embedTsnePlt(distance_threshold = 0.25,
                 sample_threshold = 5)
    twodembed_t1, sel_ids_t1 = em_ts_plt1.tsne_prune(glove_embed_load,
                                             num_points = 1000)
    words_t1 = [reverse_dictionary[i] for i in sel_ids_t1]    
    em_ts_plt1.tsne_plot(twodembed_t1, words_t1)
```
